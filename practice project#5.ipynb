```python
import pandas as pd
```

DATASET #1

```python
fish=pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Fish.csv')
```

```python
fish
```

```python
fish.columns
```

```python
#Define target X and features y
y = fish['Category']
```

```python
X = fish[['Weight', 'Height', 'Width', 'Length1',
       'Length2', 'Length3']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
```

```python
irish=pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/IRIS.csv')
```

```python
irish
```

```python
irish.columns
```

```python
#define targer X and features y
y = irish['species']
```

```python
X = irish[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
```

```python
from sklearn. model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7 , random_state = 2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
